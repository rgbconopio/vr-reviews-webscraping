```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator, get_single_color_func
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from PIL import Image
```

```python
ocquest_reviews = pd.read_csv('quest_reviews.csv')
```

```python
bad_reviews = ocquest_reviews.review[ocquest_reviews.rating <= 2]
```

```python
text = " ".join(review for review in bad_reviews).lower()
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
black = get_single_color_func('black')
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["quest", "Quest", "oculus", "VR", "the", "The", "game", 'games', 'headset','will','one','oculu','time','back'])

# Generate a word cloud image
wordcloud = WordCloud(height=2000, width=2000, background_color='white', max_words=1000, 
                      stopwords=stopwords, color_func=lambda *args, **kwargs: "black")
wordcloud.generate(text)

# Display the generated image:
# the matplotlib way:

plt.figure(figsize=[14,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file('wordcloud.png')
```

```python
logo = np.array(Image.open('oculus logo.png'))
logo.shape
```

```python
def transform_format(val):
    if val != 255:
        return 255
    else:
        return val
```

```python
transformed_logo = np.ndarray((logo.shape[0],logo.shape[1],logo.shape[2]), np.int32)
```

```python
for i in range(len(logo)):
    for z in range(len(logo[i])):
        transformed_logo[i][z] = list(map(transform_format, logo[i][z]))
```
